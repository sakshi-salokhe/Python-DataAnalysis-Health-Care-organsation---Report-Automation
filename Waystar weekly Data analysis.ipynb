```python
#importing libraries
import pandas as pd
import csv
```

```python
#file name is changed here
#paste the file path of the csv file on the local device in line number 4
# do not remove the "r" before the single quotes - File path and the file name are always in the single quotes

df = pd.read_csv(r'RAW DATA FILE.csv')

#do not change anything here onwards
df["Workgroup"].fillna(0,inplace = True)
```

```python
row_count = sum(1 for row in df["Status"])  # we calculate the total number of rows
```

```python
stat = []

#changes to be done only if any constraints are being added or any modifications are being made

for i in range(0,row_count):
    message = str(df["Last Event Message"][i])
    
    #1
    if ("claim rejected per customer request" in message.lower() and\
        df["Payer ID"][i] in ['SB801','SX073','16644','Z1093',16644]):
        stat.append("WRCM - Prof Claims that need to be archived")
        
    #2
    elif (str(df["Claim Type"][i]).lower() == "instit" and\
                                                           ("payer name matching" in message.lower() or \
                                                            "rejected per customer request" in message.lower() or\
                                                           'tddir' in message.lower())):
        stat.append("WRCM - Inst Claims that need to be archived")

    #3
    elif "smart" in message.lower() and ("supdn" in message.lower() or "uuns" in message.lower() \
                                         or "labdu" in message.lower() or "ipddn" in message.lower() \
                                         or "ahc" in message.lower() or "ccimd" in message.lower() \
                                         or "ndcdn" in message.lower() or "ubic" in message.lower()):
        stat.append("WRCM - UHC Smart Edits that need to be archived")
        
    #4
    elif "timely" in message.lower() and (df["Payer ID"][i] == 11315 or df["Payer ID"][i] == '11315'):
        stat.append("WRCM - Fidelis Timely Filing")
        
    #5
    elif (df["Payer ID"][i] == 'SX073' and "does not match allowable" in message.lower()):
        stat.append("WOPS FD - IHA missing/invalid suffix")
        
    
    #6
    elif (("patient relationship to insured is invalid" in message.lower() or \
          "subscriber id" in message.lower() or 'self' in message.lower() or 'does not match allowable pattern' in message.lower() or \
           'subscriber and subscriber id mismatched' in message.lower() or 'relationship' in message.lower()) \
          and df["Payer ID"][i] in [14165, '14165']):
        stat.append("WOPS FD - MVP/YourCare Subscriber needs to be self")

    
    #7
    elif ((df["Payer ID"][i] in ['SB804','SB805','SB806',16107,'16107','UNINW']) and \
    ("patient relationship to insured is invalid" in message.lower() or "patient relationship to insured" in message.lower() or\
    "payer requires patient information on the claim" in message.lower() or "member id not found" in message.lower() or\
    'patient billed as the' in message.lower())):
        stat.append("WOPS FD - Excellus requires Subscriber relationshi")
        
    
    #8
    elif (df["Payer ID"][i] in ['SB804','SB805','SB806','UNINW'] and ("member name" in message.lower() or "patient subscriber" in message.lower() or "subscriber id" in message.lower() or\
    "member id" in message.lower())):
        stat.append("WOPS FD - Excellus member name DOB or gender misma")
    
    
    #9
    elif (df["Payer ID"][i] == "SB807" and ("subscriber insurance type code (policy type) is missing" in message.lower() or\
                                           'insurance type code' in message.lower())):
        stat.append("WOPS FD - Medicare as Secondary Incorrect")
    
    #10    
    elif (df["Payer ID"][i] == "SB807" and ("subscriber id" in message.lower() or\
           "does not match" in message.lower() or\
           "SUBSCRIBER ANDOR SUBSCRIBER MEMBER ID NOT FOUND" in message.upper() or \
                                            "SUBSCRIBER AND OR SUBSCRIBER MEMBER ID NOT FOUND" in message.upper())):
        stat.append("WOPS FD - Medicare Eligibility")
        
        
        
    #11
    elif "excellus" not in str(df["Payer Name"][i]).lower() and ("claim has been rejected as unprocessable" in message.lower() or "Patient/Subscriber Member Name" in message or\
    "ACK/REJECT INVAL INFO" in message or\
          "incorrect payer" in message.lower() or "subscriber id is invalid" in message.lower() or \
          "member id invalid" in message.lower() or "member id not found" in message.lower() or\
          "is invalid" in message.lower() or "does not match" in message.lower() or "zip code" in message.lower() or\
          "id not found" in message.lower() or "subscriber member id" in message.lower() or\
          "subscriber id format invalid" in message.lower() or "not eligible for" in message.lower() or\
          "eligibility is not found" in message.lower() or "missing or invalid information" in message.lower() or\
          "unable to locate eligibility" in message.lower() or "claim sent to incorrect payer" in message.lower() or\
          "ssn required" in message.lower() or "patient eligibility" in message.lower() or\
          "invalid" in message.lower() or "has been canceled" in message.lower() or "mismatch" in message.lower() or \
          "relationship is invalid" in message.lower() or "must be the same as" in message.lower()) or\
    "missing/invalid member id" in message.lower() or "missing or invalid" in message.lower() or "patient id invalid" in message.lower()\
    or "entity not eligible" in message.lower():
        stat.append("WOPS FD - Subscriber Eligibility Rejections")
        
        
    #12
    elif ("CLAIM SENT TO THE INCORRECT PAYER" in message.upper() or "CLAIM SUBMITTED TO THE INCORRECT PAYER" in message.upper() or\
          "CLAIM SUBMITTED TO INCORRECT PAYER" in message.upper() or\
          "rejected per client request" in message.lower() or "CLAIM SUBMITTED TO INCORRECT PAYER" in message.upper()):
        stat.append("WOPS FD - Sent to incorrect payer")

        
    #13
    elif (df["Payer ID"][i] in ['SB807','SMNY0'] and ("facility location" in message.lower())):
        stat.append("WOPS FD - Medicare Facility Location")

        
    #14  
    elif ("procedure" in message.lower() or "electronically" in message.lower() or\
    "diagnosis" in message.lower() or "modifier" in message.lower()):
        stat.append("WRCM COD - Coding Rejections")
    
    
    #15
    elif (int(df["Sequence"][i]) == 1 and ("claim rejected per customer request" in message.lower() or \
                                     "rejected per client request" in message.lower()) and \
    df["Payer ID"][i] in ['SKNY1','SKIL0']):
        stat.append("WOPS FD - Medicaid Primary Check Elig")
    
    
    #16
    elif ((df["Payer ID"][i] == "J9899W" or df["Payer ID"][i] == "J9899A") and \
          ("auto accident" in message.lower() or 'accident date' in message.lower() or \
          'ACCEPTABLE VALUES ARE AA AUTO ACCIDENT EM EMPLOYMENT OA OTHER ACCIDENT' in message.upper() or \
          'TYPE OF ACCIDENT REQUIRED WHEN ACCIDENT DATE IS PRESENT' in message.upper())):
        stat.append("WOPS FD - Accident Type/Date invalid/missing")
    
    
    #17
    elif "national drug code (ndc)" in message.lower() or "ndc" in message.lower() or "(ndc)" in message.lower():
        stat.append("WRCM - Need NDC #")
    
    
    #18
    elif ("possible duplicated cpt codes" in message.lower() or "hcpc code s9083 on more than once" in message.lower() or \
           "duplicate of a previously processed" in message.lower()):
        stat.append("WRCM - duplicate codes")
    
    
    #19
    elif (df["Payer ID"][i] == 'SB807' and ("medicare part b payers do not accept void or replacement claims" in message.lower() or\
                                           "medicare does not accept claims with a frequency" in message.lower())):
        stat.append("WRCM - Medicare secondary need to be original")
    
    
    
    #20
    elif (int(df["Sequence"][i]) == 2 and ("out of balance" in message.lower() or\
                                     "paid amount" in message.lower() or\
                                    "please send primary payer paid adjustment amounts" in message.lower()\
                                           or "COORDINATION OF BENEFITS" in message.upper())):
        stat.append("WRCM POSTING - Claim out of balance")
        
        
        
    
    #21
    elif "the enrollment has not been completed for the submitted provider number" in message.lower() or \
    "provider not valid" in message.lower():
        stat.append("WRCM CRED - provider enrollment")
        
        
        
    #22
    elif "user has selected to discard the clam by clicking the delete button in the attachments portal" in message.lower() or \
    "user has selected to discard the claim by clicking the delete button in the attachments portal" in message.lower():
        stat.append("WRCM - WC deleted claims - need to archive")
    
    
    #23
    elif (int(df["Sequence"][i]) == 2) and str(df["Payer Name"][i]).lower() == "medicaid" and "CLAIM REJECTED PER CUSTOMER REQUEST" in message.upper():
        stat.append("WRCM - Medicaid Secondary Check primary payment")
        
        
    
    #24
    elif "cob balancing error the claim level payment" in message.lower():
        stat.append("WRCM- COB Missing or Invaild")
    
    
    elif df["Status Group"][i] == "Accepted":
        stat.append(" ")
    else:
        stat.append("WRCM - Miscellaneous")
```

```python
df["Workgroup"] = stat
```

```python
#adding new column to see if the workgroup is WOPS or WRCM
workgroup_start_word = []
for i in range(0,row_count):
    workgroups = df["Workgroup"][i].split()
    if len(workgroups) < 1:
        workgroup_start_word.append("Accepted")
    elif "WOPS" in df["Workgroup"][i]:
        workgroup_start_word.append("WOPS")
    elif "WRCM" in df["Workgroup"][i]:
        workgroup_start_word.append("WRCM")
    else:
        workgroup_start_word.append("Other")
        
df["workgroup_start_word"] = workgroup_start_word
```

```python
status1 = []
for i in range(0,row_count):
    if "rejected" in str(df["Status Group"][i]).lower():
        status1.append("Rejected")
    else:
        status1.append("Accepted")
    
df["NewStatus"] = status1
```

```python
# # drop rows with Workgroups : "WRCM - Prof Claims that need to be archived" and "WRCM - Inst Claims that need to be archived"
i = row_count-1
x = i-1
while i > 0:
    s = df["Workgroup"][i]
    
    if s == "WRCM - Prof Claims that need to be archived" or s == "WRCM - Inst Claims that need to be archived":
        df.drop(df.index[i], inplace=True)
        row_count -= 1
        i = x
        x = i-1
    else:
        i-=1
        x = i-1
if df["Workgroup"][0] == "WRCM - Prof Claims that need to be archived" or df["Workgroup"][0] == "WRCM - Inst Claims that need to be archived":
        df.drop(df.index[0], inplace=True)
        row_count -= 1
```
