```python
def bigconversion(num):
    s1 = num.split("(")
    s2 = s1[1].split(")")
    s3 = s2[0].split("$")
    #s3[1] will have the number
    s4 = s3[1].split(",")
    if len(s4) == 3:
        s5 = s4[2].split(".")
        val = s4[0]+s4[1]+s5[0]+'.'+s5[1]
    elif len(s4) == 2:
        s5 = s4[1].split(".")
        val = s4[0]+s5[0]+'.'+s5[1]
    elif len(s4) == 1:
        s5 = s4.split(".")
        val = s5[0]+'.'+s5[1]
    val = float(val)
    return val
```

```python
def conversion(num):
    validity = positive_or_negative(num)
    if validity == 1: #positive
        s3 = num.split("$")
        if "," in s3[1]:
            s4 = s3[1].split(",")
            s5 = s4[1].split(".")
            val = s4[0]+s5[0]+'.'+s5[1]
        else:
            s4 = s3[1]
            s5 = s4.split(".")
            val = s5[0]+'.'+s5[1]
        
        val = float(val)
        return val
    else: #negative
        s1 = num.split("(")
        s2 = s1[1].split(")")
        s3 = s2[0].split("$")
        if "," in s3[1]:
            s4 = s3[1].split(",")
            s5 = s4[1].split(".")
            val = s4[0]+s5[0]+'.'+s5[1]
        else:
            s4 = s3[1]
            s5 = s4.split(".")
            val = s5[0]+'.'+s5[1]

        val = -float(val)
        return val
```

```python
def positive_or_negative(num):
    if "(" in num:
        return 0
    else:
        return 1
```
